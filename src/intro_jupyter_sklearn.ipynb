```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
data = iris.data
trg = iris.target
print("#"*20)
print(iris.target_names)
```

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)

clf.fit(data[:-1], trg[:-1])

result = clf.predict(data[-1:])
print("実際の答え={0}, 予測結果={1}".format(trg[-1], result))

result = clf.score(data, trg)
print(result)
```

```python
import pickle

with open("PredictiveModel.pkl", "wb") as f:
    pickle.dump(clf, f, pickle.HIGHEST_PROTOCOL)

with open("PredictiveModel.pkl", "rb") as f:
    clf2 = pickle.load(f)

clf.predict(data) == clf2.predict(data)
```
